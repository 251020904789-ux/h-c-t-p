```python
try:
    nam = int(input("Nhập vào một năm bất kỳ: "))
    if (nam % 400 == 0) or (nam % 4 == 0 and nam % 100 != 0):
        print(f"Năm {nam} là năm nhuận.")
    else:
        print(f"Năm {nam} không phải là năm nhuận.")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ cho năm.")
```

```python
n = int(input("Nhập một số nguyên dương n: "))
tong = 0
for i in range(1, n + 1):
  tong += i
  print(f"Tổng các số từ 1 đến {n} là: {tong}")
```

```python
numbers = []
print("Nhập vào 10 số nguyên:")
for i in range(10):
    num = int(input(f"Số thứ {i+1}: "))
    numbers.append(num)
so_am = sum(1 for x in numbers if x < 0)
so_duong = sum(1 for x in numbers if x > 0)
so_0= sum(1 for x in numbers if x == 0)
print("\nKết quả:")
print(f"Số âm: {so_am}")
print(f"Số dương: {so_duong}")
print(f"Số bằng 0: {so_0}")
```

```python
import math
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
try:
    num = int(input("Nhập vào một số nguyên: "))
    if is_prime(num):
        print(f"{num} là số nguyên tố.")
    else:
        print(f"{num} không phải là số nguyên tố.")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ.")
```
