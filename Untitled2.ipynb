- `index()`: trả về kết quả là chỉ số của một phần tử trong danh sách
- `count()`: trả về kết quả là số lần xuất hiện của một phần tử trong danh sác

```python
areas = [11.25, 18.0, 20.0, 10.75, 9.50, 9.50, 11.50, 11.25]
print(areas.index(20))
print(areas.count(11.25))
```

- `append()`: thêm phần tử vào cuối danh sách được gọi
- `reverse()`: đảo ngược thứ tự các phần tử trong danh sách được gọi

```python
areas = [11.25, 18.0, 20.0, 10.75, 9.50]
areas.append(25.4)
areas.append(15.45)
print(areas)
areas.reverse()
```
