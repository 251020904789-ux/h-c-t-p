```
gpa = float(input("Nhập GPA của sinh viên (0 - 10): "))
if gpa < 0 or gpa > 10:
    print("GPA không hợp lệ!")
elif gpa >= 8.5:
    print("Học lực: Xuất sắc")
elif gpa >= 7.0:
    print("Học lực: Giỏi")
elif gpa >= 5.5:
    print("Học lực: Khá")
elif gpa >= 4.0:
    print("Học lực: Trung bình")
else:
    print("Học lực: Yếu")
```

```
n = int(input("Nhập số nguyên dương n: "))
tong = 0
for i in range(2, n + 1, 2):
    tong += i
print("Tổng các số chẵn từ 1 đến", n, "là:", tong)
```
