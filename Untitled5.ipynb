```python
import math
a = 1
b = -5
c = 6
delta = b*b - 4*a*c
if delta < 0:
    print("Phương trình vô nghiệm")
elif delta == 0:
    x = -b / (2*a)
    print("Phương trình có nghiệm kép:", x)
else:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print("Nghiệm thứ nhất:", x1)
    print("Nghiệm thứ hai:", x2)
```

```python
s = input("Enter 3 real number (separted by space):")
parts = s.split() # Corrected from fplit() and renamed to avoid confusion with original 's'

if len(parts) == 3:
    try:
        a = float(parts[0])
        b = float(parts[1])
        c = float(parts[2])
        print(f"a = {a}, b = {b}, c = {c}") # Corrected to use variables a, b, c
    except ValueError:
        print("Invalid input: Please ensure all three parts are valid numbers.")
else:
    print("Invalid input: Please enter exactly 3 numbers separated by spaces.")

# Enter your code from here
```
